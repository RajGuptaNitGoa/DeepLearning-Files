# Flatten()
In machine learning, particularly in the context of deep learning frameworks like TensorFlow/Keras and PyTorch, Flatten() is a function or layer used to transform multi-dimensional inputs into a single-dimensional (flat) vector. This process is crucial when transitioning from convolutional layers to fully connected (dense) layers in neural networks.

### Why Use Flatten()?
Transition from Convolutional to Dense Layers: Convolutional layers produce multi-dimensional outputs (feature maps) that cannot be fed directly into dense layers. Flatten() reshapes these outputs into a format that dense layers can process.
Preserve Information: While flattening changes the shape of the tensor, it preserves the information by reorganizing it into a one-dimensional array without changing the data itself.

# Dense Layers
In machine learning, particularly in the context of deep learning with neural networks, 
the Dense layer (also known as a fully connected layer) is a fundamental building block. 
It is used to create dense (fully connected) neural networks where each neuron in one layer 
is connected to every neuron in the subsequent layer.

#### Understanding Dense Layers
A Dense layer is a layer of neurons in which each neuron receives input from all the neurons in the previous layer. This layer performs a linear transformation followed by a non-linear activation function. The Dense layer is represented mathematically as:

# y=f(W*x+b)
W: Weight matrix

x: Input vector

b: Bias vector

f: Activation function (e.g., ReLU, Sigmoid, Tanh)

y: Output vector

#### Key Components
##### Weights (W): 
Each neuron has a weight that is learned during training. These weights determine the importance of the input features.
##### Bias (b): 
A bias term is added to the weighted sum of inputs, which allows the model to fit the data better by providing a degree of freedom.
##### Activation Function (f): 
An activation function introduces non-linearity to the model, allowing it to learn more complex patterns.
Common Uses of Dense Layers
##### Output Layer: 
In many neural networks, especially in classifiers, the final layer is often a dense layer. For instance, in binary classification, it might have a single neuron with a sigmoid activation function; in multi-class classification, it could have multiple neurons with a softmax activation function.
##### Hidden Layers:
Dense layers are also used as hidden layers in feedforward neural networks to capture and learn the internal representations of the data.

```python
# creating neural network
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(512, activation=tf.nn.relu),
    tf.keras.layers.Dense(10,activation=tf.nn.softmax)
])
```

Subclassing tf.keras.callbacks.Callback:
The myCallback class inherits from tf.keras.callbacks.Callback, allowing you to create a custom callback with additional functionality by overriding its methods.

Method on_epoch_end:
This method is called at the end of each epoch during training.
It takes two arguments:
epoch: An integer representing the current epoch number.
logs: A dictionary containing metrics from the current epoch. Common keys include 'accuracy', 'loss', etc.
Checking the accuracy:

logs.get('accuracy') retrieves the 'accuracy' value from the logs dictionary. This is the accuracy for the current epoch.

The if statement checks if this accuracy exceeds 0.99 (i.e., 99%).

Stopping the Training:
If the accuracy is greater than 99%, it prints a message: Reached 99% accuracy so cancelling training!.
self.model.stop_training = True sets the stop_training attribute of the model to True, which halts the training process.

an epoch refers to one complete pass through the entire training dataset

Basic Definition:

An epoch is a single iteration through the entire dataset during the training of a machine learning model. It is one cycle where the model sees all the training data once.
Epochs and Training:

During each epoch, the model processes every sample in the training dataset, calculates predictions, compares them to the actual values (using a loss function), and updates its parameters (weights and biases) to minimize the loss.
Multiple epochs are usually required to train a model effectively, as one pass through the data (one epoch) is often not enough for the model to learn patterns well.
Epochs vs. Batches:

Batch: A subset of the dataset processed at one time. If a dataset is large, it's usually split into smaller chunks or batches to fit into memory and improve computational efficiency.
Iterations: An iteration refers to one update of the model's parameters. The number of iterations per epoch is determined by the number of batches in the dataset. For example, if you have 1000 training samples and a batch size of 100, it will take 10 iterations to complete one epoch.
Mini-Batch Gradient Descent: Common training method where each batch is used to perform one update step.
Why Multiple Epochs?:

Training a model typically involves multiple epochs because each pass through the dataset helps the model adjust its weights gradually, improving performance and learning the underlying patterns more accurately over time.
More epochs mean more opportunities for the model to learn, but too many can lead to overfitting, where the model becomes too specialized to the training data and performs poorly on unseen data.
Example
Imagine a dataset with 1000 samples. If you train a model with a batch size of 100 and set it to train for 5 epochs:

Each epoch will process all 1000 samples.
Each epoch will be broken down into 10 batches (since 1000 / 100 = 10).
The model will update its weights 10 times per epoch.
After 5 epochs, the model will have seen each sample 5 times (in different contexts, as batches might shuffle the data).

```python
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if(logs.get('accuracy')>0.99):
            print("\nReached 99% accurracy so cancelling training!")
            self.model.stop_training=True
```

```python
callbacks=myCallback()
```

```python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100,callbacks=[callbacks])
```

```python
model.evaluate(x_test,y_test)
```

```python
def preprocess_image(img_path):
    img = Image.open(img_path)  # Convert to grayscale
    img = img.resize((28, 28))  # Resize to 28x28 pixels
    img = np.array(img)  # Convert to numpy array
    img = img / 255.0  # Normalize pixel values
    img = np.reshape(img, (1, 28, 28, 1))  # Reshape to (1, 28, 28, 1)
    return img
```

```python
# Function to predict the digit from an image
def predict_digit(img_path):
    processed_image = preprocess_image(img_path)
    prediction = model.predict(processed_image)
    predicted_digit = np.argmax(prediction, axis=1)[0]
    return predicted_digit

```

```python
# Example usage: Predicting from an image file
img_path = 'i2.png'  # Update with your image path
predicted_digit = predict_digit(img_path)
print(f'Predicted Digit: {predicted_digit}')
```
